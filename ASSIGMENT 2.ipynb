```python
mean = data["Speed"].mean()
print(mean)
def sp(val):
    if val > mean:
        return "High_speed"
    else:
        return "Low_speed"
```

```python
data["High_Low_Speed"] = data["Speed"].apply(sp)
```

```python
data.head()
```

```python
mean = data["HP"].mean()
print(mean)
def Hp(val):
    if val > mean:
        return "High_HP"
    else:
        return "Low_HP"
```

```python
data["High_Low_HP"] = data["HP"].apply(Hp)
```
